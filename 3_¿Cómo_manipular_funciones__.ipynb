## Dependencias

```python
import numpy as np
import matplotlib.pyplot as plt 
```

## Desplazamientos verticales y horizontales

Siendo $c$ una constante mayor que cero, entonces la gr√°fica:

- $y=f(x)+c$ se desplaza $c$ unidades hacia arriba.
- $y=f(x)-c$ se desplaza $c$ unidades hacia abajo.
- $y=f(x-c)$ se desplaza $c$ unidades hacia la derecha.
- $y=f(x+c)$ se desplaza $c$ unidades hacia la izquierda.

```python
N = 1000

def f(x):
  return x**2;

c = 4

x = np.linspace(-10,10, num=N)

y = f(x + c)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

## Alargamientos y compresiones

Siendo $c$ una constante mayor que cero, entonces la gr√°fica:

- $y=c \cdot f(x)$ alarga la gr√°fica verticalmente en un factor de $c$.
- $y= \frac{1}{c} \cdot f(x)$ comprime la gr√°fica verticalmente en un factor de $c$.
- $y=f(c \cdot x)$ comprime la gr√°fica horizontelmente en un factor de $c$.
- $y= f(\frac{1}{c} \cdot x )$ alarga la gr√°fica horizontelmente en un factor de $c$.

```python
N = 1000

def f(x):
  return np.sin(x);

c = 2

x = np.linspace(-15,15, num=N)

y = f((1/c)*x)

z = f(c*x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.plot(x,z)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

## Reflexiones

- $y=-f(x)$ refleja la gr√°fica respecto al eje x.
- $y=f(-x)$ refleja la gr√°fica respecto al eje y.

```python
N = 1000

def f(x):
  return x**3;

x = np.linspace(-10,10, num=N)

y = f(-x)

z = f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.plot(x,z)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

# Caracter√≠sticas de las funciones
Se les llama funciones reales porque tanto su dominio como el codominio (recuerda que al codominio tambi√©n se le puede llamar rango o imagen) est√°n contenidos en el conjunto de los n√∫meros reales. Es decir, el conjunto que contiene a los n√∫meros racionales e irracionales. En otras palabras cualquier n√∫mero que se te ocurra que no sea imaginario. Todos los ejemplos que veamos a lo largo de este curso ser√°n sobre n√∫meros reales.

Una vez que ya tenemos claro cu√°les son las funciones reales y por qu√© se les llama as√≠, pasemos a describir algunas de sus caracter√≠sticas.

## Funci√≥n par

Una funci√≥n es par si cumple la siguiente relaci√≥n a lo largo de su dominio:

$$f(-x)=f(x)$$

Si lo notaste, esta relaci√≥n nos dice que una funci√≥n es par si es sim√©trica al eje vertical (eje Y). Por ejemplo, una par√°bola es una funci√≥n es par.

## Funci√≥n impar

Una funci√≥n es impar si cumple la siguiente relaci√≥n a lo largo de su dominio:

$$f(-x)=-f(x)$$

Esta relaci√≥n nos indica que una funci√≥n es impar si es sim√©trica al eje horizontal (eje X). Por ejemplo, una funci√≥n c√∫bica es impar.

Las dos caracter√≠sticas o propiedades anteriores est√°n muy relacionadas con la simetr√≠a. Puedes comprobarlas t√∫ mismo con diferentes funciones e incluso hacer una peque√±a rutina en Python que compruebe si una funci√≥n es par o impar.

## Funci√≥n acotada

Una funci√≥n es acotada si su codominio (tambi√©n conocido como rango o imagen) se encuentra entre dos valores, es decir, est√° acotado. Esta definici√≥n se define como que hay un n√∫mero m que para todo valor del dominio de la funci√≥n se cumple que:

$$-m ‚â§ f(x) ‚â§ m$$

Por ejemplo, la funci√≥n seno o coseno est√°n acotadas en el intervalo [-1, 1] dentro de su co-dominio.

## Funciones Mon√≥tonas

Estas funciones son √∫tiles de reconocer o analizar debido a que nos permiten saber si una funci√≥n crece o decrece en alguno de sus intervalos. Que algo sea mon√≥tono significa que no tiene variaciones. Entonces las funciones mon√≥tonas son aquellas que dentro de un intervalo I, perteneciente a los n√∫meros reales, cumple alguna de estas propiedades:

1. La funci√≥n es mon√≥tona y estrictamente creciente:

si para todo:
$$ x_{1} , x_{2} ‚àà I: x_{1} < x_{2} \Longleftrightarrow f(x_{1}) < f(x_{2})$$

La forma correcta de leer esto es ‚Äúsi para todo x1 y x2 que pertenecen al intervalo I, tal que x1 sea menor a x2, si y solo si f(x1) sea menor a f(x2)‚Äù. En palabras mucho m√°s sencillas, lo que nos dice esta definici√≥n es que x1 siempre tiene que ser menor que x2 en nuestro intervalo I, y que al evaluar x2 en la funci√≥n el resultado de esto siempre ser√° mayor que si evaluamos la funci√≥n en x1. Para las siguientes tres definiciones restantes no cambia mucho la forma en la que se interpretan.

2. La funci√≥n es mon√≥tona y estrictamente decreciente:

si para todo:
$$ x_{1} , x_{2} ‚àà I: x_{1} < x_{2} \Longleftrightarrow f(x_{1}) > f(x_{2})$$

3. La funci√≥n es mon√≥tona y creciente

si para todo:
$$ x_{1} , x_{2} ‚àà I: x_{1} < x_{2} \Longleftrightarrow f(x_{1}) ‚â§ f(x_{2})$$

4. La funci√≥n es mon√≥tona y decreciente

si para todo:
$$ x_{1} , x_{2} ‚àà I: x_{1} < x_{2} \Longleftrightarrow f(x_{1}) \ge f(x_{2})$$

## Funciones Periodicas

Las funciones peri√≥dicas son aquellas que se repiten cada cierto periodo, este periodo se denomina con la letra T. La relaci√≥n que debe cumplir la funci√≥n para ser peri√≥dica es la siguiente.

$$f(x) = f(x+T), T \neq 0$$

Por ejemplo, la funci√≥n seno y coseno son funciones peri√≥dicas con un periodo T = 2œÄ. Es decir que si nosotros calculamos f(x) y calculamos f(x + 2œÄ) en la funci√≥n seno el valor que nos den ambas expresiones es el mismo.
